### 2. Load data

```python
file_path = r"C:\Users\subhi.gupta\Downloads\Bike Sharing\datasets\train.csv"
df = pd.read_csv(file_path,  parse_dates=["datetime"])
df
```

### 3. Create new features

```python
df['year'] = df['datetime'].dt.year
df['month'] = df['datetime'].dt.month
df['day'] = df['datetime'].dt.day
df['weekday'] = df['datetime'].dt.dayofweek
df['hour'] = df['datetime'].dt.hour
```

### 4. Heatmaps

```python
new_df = df.copy()
new_df.drop("datetime", axis=1, inplace=True)
```

```python
new_df
```

```python
corr_matrix = new_df.corr()

plt.figure(figsize=(14, 10))
sns.heatmap(
    corr_matrix,
    cmap='coolwarm',
    annot=True,      # use True if you want numbers shown
    fmt='.2f',
    linewidths=0.5
)
plt.title('Feature Correlation Heatmap', fontsize=16)
plt.show()
```

```python
price_corr = corr_matrix['count'].abs().sort_values(ascending=False)
for feature, corr in price_corr.items():
    if feature != 'price':
        print(f"{feature:20s}: {corr:6.3f}")
```
